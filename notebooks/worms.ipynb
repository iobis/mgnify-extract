# WoRMS matching notes

Get Darwin Core tables for a single study:

```python
from mgnifyextract.dwc import study_to_dwc
from mgnifyextract.studies import get_superstudy_studies

study = get_superstudy_studies("atlanteco")[0]
occ, dna = study_to_dwc(study, max_samples=1, markers=["LSU", "SSU"], add_lsid=True)
```

Select distinct taxonomic assignments:

```python
assignments = occ.filter(items=["scientificName", "taxonomy_id", "scientificNameID"]).drop_duplicates().sort_values(by="scientificNameID")
assignments
```

Missing LSID but NCBI ID exists:

```python
assignments[assignments["scientificNameID"].isna() & assignments["taxonomy_id"]]
```

Missing NCBI taxonomy ID:

```python
assignments[assignments["taxonomy_id"].isna()]
```
